```python
# Dropping duplicates
df = df.drop_duplicates()

# Removing NaN values
df = df.dropna()
```

```python
df
```

```python
# Showing a corelation map for all numeric variables
corr_matrix = df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('corelation Map for Numeric variables')
plt.show()
```

# Visualize the categorical variables

```python
data=df
```

```python
# overtime
sns.countplot(data=df, x='OverTime')
plt.title('OverTime')
plt.show()
```

```python
# Marital  status
sns.countplot(data= df, x='MaritalStatus')
plt.title('Marital Status')
plt.show()
```

```python
# Job Role
plt.figure(figsize =(10,5))
sns.countplot(data= df, y = 'JobRole')
plt.title('Job Role')
plt.show()
```

```python
# Gender
sns.countplot(data= df, x ='Gender')
plt.title('Gender')
plt.show()
```

```python
# Education Field
plt.figure(figsize=(10,5))
sns.countplot(data=df, y = 'EducationField')
plt.title('Education Field')
plt.show()
```

```python
# Department
sns.countplot(data=df, x='Department')
plt.show()
```

```python
# Business Travel
sns.countplot(data=df, x='BusinessTravel')
plt.title('Business Travel')
plt.show()
```

```python
# Relation Between Overtime and age
sns.boxplot(data=df, x='OverTime', y='Age') 
plt.title('Relationship Between Overtime and Age')
plt.show()
```

# Plotting numerical variables

```python
# Total working Years 
sns.histplot(data=df, x='TotalWorkingYears', bins=10, kde=True)
plt.title('TotalworkingYears') 
plt.show()
```

```python
#Number of Companies Worked
sns.histplot(data= df, x='NumCompaniesWorked', bins=10, kde=True)
plt.title('Number of Companies Worked')
plt.show()
```

```python
# Distance From Home
plt.title('Distance From Home')
sns.histplot(data=df, x='DistanceFromHome', bins=10, kde = True)
plt.show()
```

# Salary Analysis

```python
# Monthly Income Distribution
sns.histplot(data=df, x='MonthlyIncome', bins = 10, kde=True)
plt.show()
```

```python
# salary by job role
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='JobRole', y='MonthlyIncome')
plt.title('Salary by Job Role')
plt.xticks(rotation = 90)
plt.show()
```
